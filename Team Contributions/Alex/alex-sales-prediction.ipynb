```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import & read CSV file

csv_data = 'data.csv'


# Create dataframe based on CSV file

data = pd.read_csv(csv_data, encoding='latin1')

list(data.columns)
```

```python
data.head()
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['InvoiceDate'])

data.head()
```

```python
data['day'] = data.Date.dt.day

data['month'] = data.Date.dt.month

data['year'] = data.Date.dt.year

data
```

```python
data['ext_price'] = data['Quantity'] * data['UnitPrice']

data
```

```python
uk = data.loc[data["Country"] == "United Kingdom", :]
uk
```

```python
uk_2010 = uk.loc[uk["year"] == 2010, :]
uk_2010.head()
```

```python
uk_2010.info()
```

```python
invoices = len(uk_2010["InvoiceNo"].unique())
invoices
```

```python
unique_items = len(uk_2010["StockCode"].unique())
unique_items
```

```python
total_items = uk_2010["Quantity"].sum()
total_items
```

```python
total_sales = uk_2010["ext_price"].sum()
total_sales
```

```python
df1 = pd.DataFrame({'total_invoices' : [invoices],
                    'unique_products' : [unique_items],
                    'products_sold' : [total_items],
                    'total_sales' : [total_sales]})
df1.head()
```

```python
uk_2011 = uk.loc[uk["year"] != 2010, :]
uk_2011_Jan = uk_2011.loc[uk_2011["month"] == 1, :]
uk_2011_Feb = uk_2011.loc[uk_2011["month"] == 2, :]
uk_2011_Mar = uk_2011.loc[uk_2011["month"] == 3, :]
uk_2011_Apr = uk_2011.loc[uk_2011["month"] == 4, :]
uk_2011_May = uk_2011.loc[uk_2011["month"] == 5, :]
uk_2011_Jun = uk_2011.loc[uk_2011["month"] == 6, :]
uk_2011_Jul = uk_2011.loc[uk_2011["month"] == 7, :]
uk_2011_Aug = uk_2011.loc[uk_2011["month"] == 8, :]
uk_2011_Sep = uk_2011.loc[uk_2011["month"] == 9, :]
uk_2011_Oct = uk_2011.loc[uk_2011["month"] == 10, :]
uk_2011_Nov = uk_2011.loc[uk_2011["month"] == 11, :]
uk_2011_Dec = uk_2011.loc[uk_2011["month"] == 12, :]
```

```python
uk_2011_Jan

invoices = len(uk_2011_Jan["InvoiceNo"].unique())

unique_items = len(uk_2011_Jan["StockCode"].unique())

total_items = uk_2011_Jan["Quantity"].sum()

total_sales = uk_2011_Jan["ext_price"].sum()

df2 = pd.DataFrame({'total_invoices' : [invoices],
                    'unique_products' : [unique_items],
                    'products_sold' : [total_items],
                    'total_sales' : [total_sales]})
df2.head()
```

```python
points = [uk_2011_Jan, uk_2011_Feb, uk_2011_Mar, uk_2011_Apr, uk_2011_May, uk_2011_Jun, uk_2011_Jul, uk_2011_Aug, uk_2011_Sep, uk_2011_Oct, uk_2011_Nov, uk_2011_Dec] 

invoices = []

unique_items = []

total_items = []

total_sales = []

for i in points:
    
    invoices.append(len(i["InvoiceNo"].unique()))

    unique_items.append(len(i["StockCode"].unique()))

    total_items.append(i["Quantity"].sum())

    total_sales.append(i["ext_price"].sum())
    
```

```python
uk_df = pd.DataFrame(list(zip(invoices, unique_items, total_items, total_sales)),
                     columns = ['total_invoices', 'unique_products', 'products_sold', 'total_sales'])

uk_df
```

```python
data_frame = pd.concat([df1, uk_df])
data_frame
```
